# Região Centro Oeste

```python
## Lendo o arquivo do Centro Oeste do ano 2014 com as informações necessárias para análise
df =pd.read_csv(r'CENTRO_OESTE_2014.csv', delimiter='|', encoding='iso-8859-1', usecols=['ID_TIPO_NEC_ESP_AUTISMO', 'FK_COD_ETAPA_ENSINO','ID_DEPENDENCIA_ADM_ESC'])

## Contando as informações necessárias para análise
## ID_TIPO_NEC_ESP_AUTISMO - Identificando se a criança matriculado tem o diagnóstico de Autismo
## FK_COD_ETAPA_ENSINO - se a criança matriculada está na etapa inicial de ensino do 1ª ao 5ª ano
## ID_DEPENDENCIA_ADM_ESC - se é escola pública ou privada

## Guarda a contagem como lista, sendo na primeira posição escola pública[0], na segunda posição escola privada[1]
co14.append(df.query('(ID_TIPO_NEC_ESP_AUTISMO == 1) and (FK_COD_ETAPA_ENSINO>=14)and(FK_COD_ETAPA_ENSINO<=18 and ID_DEPENDENCIA_ADM_ESC != 4)')['ID_TIPO_NEC_ESP_AUTISMO'].count())
co14.append(df.query('(ID_TIPO_NEC_ESP_AUTISMO == 1) and (FK_COD_ETAPA_ENSINO>=14)and(FK_COD_ETAPA_ENSINO<=18 and ID_DEPENDENCIA_ADM_ESC == 4)')['ID_TIPO_NEC_ESP_AUTISMO'].count())


## Lendo o arquivo do Centro Oeste do ano 2015 com as informações necessárias para análise
df = pd.read_csv(r'CENTRO_OESTE_2015.csv', delimiter='|', encoding='iso-8859-1', usecols=['IN_AUTISMO', 'TP_ETAPA_ENSINO','TP_DEPENDENCIA'])
## Contando as informações necessárias para análise
co15.append(df.query('(IN_AUTISMO == 1) and (TP_ETAPA_ENSINO>=14)and(TP_ETAPA_ENSINO<=18) and TP_DEPENDENCIA != 4')['IN_AUTISMO'].count())
co15.append(df.query('(IN_AUTISMO == 1) and (TP_ETAPA_ENSINO>=14)and(TP_ETAPA_ENSINO<=18) and TP_DEPENDENCIA == 4')['IN_AUTISMO'].count())


## Lendo o arquivo do Centro Oeste do ano 2016 com as informações necessárias para análise
df = pd.read_csv(r'CENTRO_OESTE_2016.csv', delimiter='|', encoding='iso-8859-1', usecols=['IN_AUTISMO', 'TP_ETAPA_ENSINO','TP_DEPENDENCIA'])
## Contando as informações necessárias para análise
co16.append(df.query('(IN_AUTISMO == 1) and (TP_ETAPA_ENSINO>=14)and(TP_ETAPA_ENSINO<=18) and TP_DEPENDENCIA != 4')['IN_AUTISMO'].count())
co16.append(df.query('(IN_AUTISMO == 1) and (TP_ETAPA_ENSINO>=14)and(TP_ETAPA_ENSINO<=18) and TP_DEPENDENCIA == 4')['IN_AUTISMO'].count())


## Lendo o arquivo do Centro Oeste do ano 2017 com as informações necessárias para análise
df = pd.read_csv(r'CENTRO_OESTE_2017.csv', delimiter='|', encoding='iso-8859-1', usecols=['IN_AUTISMO', 'TP_ETAPA_ENSINO','TP_DEPENDENCIA'])
## Contando as informações necessárias para análise
co17.append(df.query('(IN_AUTISMO == 1) and (TP_ETAPA_ENSINO>=14)and(TP_ETAPA_ENSINO<=18) and TP_DEPENDENCIA != 4')['IN_AUTISMO'].count())
co17.append(df.query('(IN_AUTISMO == 1) and (TP_ETAPA_ENSINO>=14)and(TP_ETAPA_ENSINO<=18) and TP_DEPENDENCIA == 4')['IN_AUTISMO'].count())


## Lendo o arquivo do Centro Oeste do ano 2018 com as informações necessárias para análise
df = pd.read_csv(r'CENTRO_OESTE_2018.csv', delimiter='|', encoding='iso-8859-1', usecols=['IN_AUTISMO', 'TP_ETAPA_ENSINO','TP_DEPENDENCIA'])
## Contando as informações necessárias para análise
co18.append(df.query('(IN_AUTISMO == 1) and (TP_ETAPA_ENSINO>=14)and(TP_ETAPA_ENSINO<=18) and TP_DEPENDENCIA != 4')['IN_AUTISMO'].count())
co18.append(df.query('(IN_AUTISMO == 1) and (TP_ETAPA_ENSINO>=14)and(TP_ETAPA_ENSINO<=18) and TP_DEPENDENCIA == 4')['IN_AUTISMO'].count())

```

```python
data = [[2014, 2014, 2015, 2015, 2016, 2016, 2017, 2017, 2018, 2018],
        ['Pública', 'Privada', 'Pública', 'Privada', 'Pública', 'Privada', 'Pública', 'Privada', 'Pública','Privada',],
        [co14[0], co14[1], co15[0], co15[1], co16[0], co16[1], co17[0], co17[1], co18[0], co18[1]],
       10*['Centro-Oeste']]

rows = zip(data[0], data[1], data[2], data[3])
headers = ['Ano', 'Dependência Administrativa', 'Número de Matrículas','Região']
df1 = pd.DataFrame(rows, columns=headers)
df1
```

```python
df1.groupby(['Ano','Dependência Administrativa'])['Número de Matrículas'].sum().unstack(1).plot.bar(figsize=(15,8));
plt.title('Evolução das matrículas de crianças diagnosticadas com Autismo nos anos iniciais - Região Centro Oeste', fontsize=16)
plt.xlabel('Ano',fontsize=14)
plt.ylabel('Número de matriculas',fontsize=14)
    
```

```python
## Lendo o arquivo da Região do Sul do ano 2014 com as informações necessárias para análise
df =pd.read_csv(r'SUL_2014.csv', delimiter='|', encoding='iso-8859-1', usecols=['ID_TIPO_NEC_ESP_AUTISMO', 'FK_COD_ETAPA_ENSINO','ID_DEPENDENCIA_ADM_ESC'])

## Contando as informações necessárias para análise
## ID_TIPO_NEC_ESP_AUTISMO - Identificando se a criança matriculado tem o diagnóstico de Autismo
## FK_COD_ETAPA_ENSINO - se a criança matriculada está na etapa inicial de ensino do 1ª ao 5ª ano
## ID_DEPENDENCIA_ADM_ESC - se é escola pública ou privada
## Guarda a contagem como lista, sendo na primeira posição escola pública[0], na segunda posição escola privada[1]
s14.append(df.query('(ID_TIPO_NEC_ESP_AUTISMO == 1) and (FK_COD_ETAPA_ENSINO>=14)and(FK_COD_ETAPA_ENSINO<=18 and ID_DEPENDENCIA_ADM_ESC != 4)')['ID_TIPO_NEC_ESP_AUTISMO'].count())
s14.append(df.query('(ID_TIPO_NEC_ESP_AUTISMO == 1) and (FK_COD_ETAPA_ENSINO>=14)and(FK_COD_ETAPA_ENSINO<=18 and ID_DEPENDENCIA_ADM_ESC == 4)')['ID_TIPO_NEC_ESP_AUTISMO'].count())

## Lendo o arquivo da região do Sul do ano 2015 com as informações necessárias para análise
df = pd.read_csv(r'SUL_2015.csv', delimiter='|', encoding='iso-8859-1', usecols=['IN_AUTISMO', 'TP_ETAPA_ENSINO','TP_DEPENDENCIA'])
## Contando as informações necessárias para análise
s15.append(df.query('(IN_AUTISMO == 1) and (TP_ETAPA_ENSINO>=14)and(TP_ETAPA_ENSINO<=18) and TP_DEPENDENCIA != 4')['IN_AUTISMO'].count())
s15.append(df.query('(IN_AUTISMO == 1) and (TP_ETAPA_ENSINO>=14)and(TP_ETAPA_ENSINO<=18) and TP_DEPENDENCIA == 4')['IN_AUTISMO'].count())

## Lendo o arquivo da região do Sul do ano 2016 com as informações necessárias para análise
df = pd.read_csv(r'SUL_2016.csv', delimiter='|', encoding='iso-8859-1', usecols=['IN_AUTISMO', 'TP_ETAPA_ENSINO','TP_DEPENDENCIA'])
## Contando as informações necessárias para análise
s16.append(df.query('(IN_AUTISMO == 1) and (TP_ETAPA_ENSINO>=14)and(TP_ETAPA_ENSINO<=18) and TP_DEPENDENCIA != 4')['IN_AUTISMO'].count())
s16.append(df.query('(IN_AUTISMO == 1) and (TP_ETAPA_ENSINO>=14)and(TP_ETAPA_ENSINO<=18) and TP_DEPENDENCIA == 4')['IN_AUTISMO'].count())

## Lendo o arquivo da região do Sul do ano 2017 com as informações necessárias para análise
df = pd.read_csv(r'SUL_2017.csv', delimiter='|', encoding='iso-8859-1', usecols=['IN_AUTISMO', 'TP_ETAPA_ENSINO','TP_DEPENDENCIA'])
## Contando as informações necessárias para análise
s17.append(df.query('(IN_AUTISMO == 1) and (TP_ETAPA_ENSINO>=14)and(TP_ETAPA_ENSINO<=18) and TP_DEPENDENCIA != 4')['IN_AUTISMO'].count())
s17.append(df.query('(IN_AUTISMO == 1) and (TP_ETAPA_ENSINO>=14)and(TP_ETAPA_ENSINO<=18) and TP_DEPENDENCIA == 4')['IN_AUTISMO'].count())

## Lendo o arquivo da região do Sul do ano 2018 com as informações necessárias para análise
df = pd.read_csv(r'SUL_2018.csv', delimiter='|', encoding='iso-8859-1', usecols=['IN_AUTISMO', 'TP_ETAPA_ENSINO','TP_DEPENDENCIA'])

## Contando as informações necessárias para análise
s18.append(df.query('(IN_AUTISMO == 1) and (TP_ETAPA_ENSINO>=14)and(TP_ETAPA_ENSINO<=18) and TP_DEPENDENCIA != 4')['IN_AUTISMO'].count())
s18.append(df.query('(IN_AUTISMO == 1) and (TP_ETAPA_ENSINO>=14)and(TP_ETAPA_ENSINO<=18) and TP_DEPENDENCIA == 4')['IN_AUTISMO'].count())
```
